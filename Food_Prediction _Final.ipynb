```python
def CountFrequency(my_list): 
  
    # Creating an empty dictionary  
    freq = {} 
    for item in my_list: 
        if (item.strip() in freq): 
            freq[item.strip()] += 1
        else: 
            freq[item.strip()] = 1
  
    for key, value in freq.items(): 
        return(freq)
```
